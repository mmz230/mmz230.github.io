## Data Types
1. **Structured data:** highly organized information that can be seamlessly included in a database and readily searched via simple search operations
2. **Unstructured data:** it is essentially the opposite of  structured data, devoid of any underlying structure

Roughly speaking, a structured data contains labels or fields, which tells us what the number, text or category means. In this course, we will be mostly using structured data.

## Data Acquisition 

1. *Acquisition: act of acquiring something.*
2. Data Acquisition: acquiring data for analysis. 
3. It could be done via:
 - **Direct approach:** is a process that converts analog and/or real world data into digital data that can be processed for further analysis. Data Acquisition System (DAS) are used in this acquisition.
 - **Indirect approach:** is a process of identifying sources/vendors that provides the data based on specifications.  

## Data Collections
1. Collections of data available for analysis. For example: Open data repositories and Social media data. 
2. Legality and usage rights should be considered before using data from collections.

## Data Storage

Structured text data (*the data we consider in this course*) can be stored in the following formats:
1. Comma Separated Values **CSV**  
2. Tab Separated Values **TSV**
3. eXtensible Markup Language **XML**
4. Really Simple Syndication **RSS**
5. JavaScript Object Notation **JSON**  


We will be mostly working with the CSV or TSV files in this course. However, data in any of the above forms can be read into Python (or any programming language of your choice) using standard procedures. For example, to read:
- CSV, TSV: pandas.read_csv
- JSON: pandas.read_json

# Data Preprocessing
> ðŸ“Œ Real world data is often *dirty* and needs preprocessing before its desired usage.

What makes data *dirty*:
1. Incomplete
2. Noise
3. Inconsistency 

## Data Scrubbing/Cleaning

Following are different ways to clean data:

1. Handling missing data: Global constant, Ignoring, Imputing or Inference approach.
2. Filtering noisy data: Identifying outliers and removing from database.
3. Munging or Wrangling: Involves reformatting data into usable format.

## Case Study-1

Consider the data given in Dataset-1.csv file. For this case study, do the following:
1. Read and display the data using Python pandas library.
2. Identify the fields of the data.
3. Identify the type for each field based on value. Also, identify the datatypes in Python. Resolve any inconsistency.
4. Clean the data using all the applicable techniques mentioned below:
   - Filter/smooth noisy data: Any negative numeric value should be replaced by its absolute value.
   - Handle any missing data appropriately: Replace it with mean or mode for numeric or non-numeric data respectively.
   - Do data wrangling/mugging: One of the data collectors entered life expectancies in fraction instead of percentages
5.  Display the transformed Dataset

```python
#1. Reading & displaying the data 
import pandas as pd
df = pd.read_csv('data/Dataset-1.csv', delimiter = ',')
display(df.head())
#display(df)
```

2. Identify the fields of the data.  
From the above table, we can see the following fields: *Country, Population, Overall Life, Male Life, Female Life, Continent, and Subregion.

3. Identify the type for each field based on value. Also, identify the datatypes in Python. Resolve any inconsistency.
The following table gives the required information:

|     Field    |     Type    |
|:------------:|:-----------:|
|    Country   | Categorical |
|  Population  |   Numeric   |
| Overall Life |   Numeric   |
|   Male Life  |   Numeric   |
|  Female Life |   Numeric   |
|   Continent  | Categorical |
|   Subregion  | Categorical |


```python
# datatypes in python
display(df.info())

# resolving inconsistency
df["Population"]=df["Population"].apply(lambda x: int(x.replace(",","")))
display(df.info())
```

```python
#4. Clean the data using all the applicable techniques mentioned below:

# Filter/smooth noisy data: Any negative numeric value should be replaced by its absolute value.
#display(df)

# In row 7 (index 6) under field Male Life, there is a negative value. 
# Since expectancy cannot be negative, this is a noisy value. 
# Looking at the Overall and Female Fields, clearly the minus sign is a typo 
# that must have been entered while entering the data. To remove the typo, we can do the following:
# df.iloc[6,3] = -df.iloc[6,3]
# However, in a typical data set, the exact location(s) may not be readily available. So we use the following:

seleted_cols= df.select_dtypes(exclude='object').columns
df[seleted_cols]=df[seleted_cols].applymap(lambda x: x if x>=0 else abs(x))
display(df)

```

```python
# Handle any missing data appropriately: Replace it with mean or mode for numeric or non-numeric data respectively.

# In row 6 (index 5) under field Overall Life, there is a NaN value 
# (NaN stands for not a number, it means there is no value here). 
# Looking at the Male and Female Fields, we can estimate the missing value
# To fill the missing value, we can do the following:
# df.iloc[5,2] = df.iloc[:,2].mean()
# However, in a typical data set, the exact location(s) may not be readily available. So we use the following:



# # To identity the location of missing values we can use the following:
#df.isna()


#  Identify the columns containing null values
null_columns=df.columns[df.isna().any()]
print('Columns with NaN values are:', null_columns)

#fill NaN values with mean or mode (only works for single missing value in categorical variable)
#for c in null_columns:
#   value = df[c].mean() if df[c].dtype!='object' else df[c].mode()
#   df[c].fillna(value,inplace=True)
    
#fill NaN values with mean or mode (modified)
for c in null_columns:
    if df[c].dtype!='object':
        value = df[c].mean()
    else:
        value = df[c].mode()
        #print (value)
        value = value[0]  #0 will be the row(s) name
        #print (value)
    df[c].fillna(value,inplace=True)

display(df)

 
    
    


## fill NaN values with mean or mode without loop.(only works for single missing value in categorical variable)
#df['Overall Life'].fillna(value=df['Overall Life'].mean())
#df['Continent'].fillna(value=df['Continent'].mode())

# # Here we filled NaN or missing values with mean or mode. 
# # However, we can flill them with any other suitable value based on the context.
# # Or, we can drop the missing values as follows:
# #     We can drop the rows containing missing values.
# #     We can drop the columns containing missing values.

# # To drop all columns containing 20 or more missing values. 
# # That is, keep only the columns with df.shape[0]-19 or more NON-NA values.
# df.dropna(axis=1, how='any', thresh=df.shape[0]-19, inplace=True)


# To drop all rows containing all NaN values.
# df.dropna(axis=0, how='all', inplace=True)
```

```python
# Do data wrangling/mugging: One of the data collectors entered life expectancies in fraction instead of percentages

# In row 11 (index 10) all the numeric values are in fraction instead of percentages. 
# To reformat the values, we can do the following:
# df.iloc[10,2:5] = 100*df.iloc[10,2:5]
# However, in a typical data set, the exact location(s) may not be readily available. So we use the following:
seleted_cols = df.loc[:,"Overall Life":"Female Life"].columns
df[seleted_cols]=df[seleted_cols].applymap(lambda x: x if x>1 else 100*x)


#5.  Display the transformed Dataset
display(df)
```

## Case Study-2

Consider the data given in Dataset-2.csv file. For this case study, do the following:

1. Read the data and display the fields (and data) of the data using Python Pandas library.
2. Identify the an ordinal field containing 3 or more categories.
3. For *education* field, convert the categories to corresponding numerical values as follows: 
  -Convert 'primary' to 1.
  -Convert 'secondary' to 2.
  -Convert 'tertiary' to 3.
  -delete all records with 'unknown' value.

4. Pick field *age*, and normalize it such that the max value is 1 and min value is 0.
5. Pick field *balance*, and normalize it such that the mean is zero, and standard deviation is 1.
6. Pick field *subscribed*, and label 'Yes' as 1, and 'No' as 0.
7. Pick field *marital*, and do one hot encoding. Remove the original column and remove the first encoded column.
8. Pick fields *default, housing, loan*, and do one hot encoding. Remove the original column.
9. Pick field *job*, and discretize the categories as follows:
    - 'unemployed':'C1', 'self-employed':'C1', 'student':'C1', 'retired':'C1'
    - 'services':'C2', 'blue-collar':'C2', 'technician':'C2', 'housemaid':'C2'
    - 'management':'C3', 'entrepreneur':'C3', 'admin.':'C3'
    - delete all records with 'unknown' value.  
    Then do one hot encoding and remove the original column.
10. Display the transformed dataset.

```python
#1. Read the data and display the fields (and data) of the data using Python Pandas library.

import pandas as pd
# import numpy as np
df = pd.read_csv('data/Dataset-2.csv', delimiter = ',')

display(df.info())
display(df.head())
```

```python
#2. Identify the an ordinal field containing 3 or more categories.

# Display unique values for non-numeric fields
print(df.select_dtypes("object").describe().loc['unique', :])

# It looks like only education is the ordinal field with more than 3 categories.
# The categories are:
df['education'].unique().tolist()
```

```python
#3. For *education* field, convert the categories to corresponding numerical values as follows: 
#   -Convert 'primary' to 1.
#   -Convert 'secondary' to 2.
#   -Convert 'tertiary' to 3.
#   -delete all records with 'unknown' value.
    
# row indicies which has unknown value for education
index_row = df[ df['education'] == 'unknown' ].index

# Delete these row indexes from dataFrame
df.drop(index_row , inplace=True)
uniqueEducation=df['education'].unique().tolist()
print(uniqueEducation)
```

```python
# edu_mapping = dict(zip(uniqueEducation,list(range(1,4))))

## Or make the dictionary manually
edu_mapping = {'primary':1, 
               'secondary':2, 
               'tertiary':3}

df['education'] = df['education'].map(edu_mapping)
display(df)
```

```python
#4. Pick field *age*, and normalize it such that the max value is 1 and min value is 0.

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['age']])
df['age']=scaler.transform(df[['age']])

display(df)
print(f'The max and min values are', df['age'].max(), ' and ', df['age'].min())
```

```python
#5. Pick field *balance*, and normalize it such that the mean is zero, and standard deviation is 1.

from sklearn.preprocessing import StandardScaler
scaler2 = StandardScaler()
scaler2.fit(df[['balance']])
df['balance']=scaler2.transform(df[['balance']])

display(df)
print(f'The mean and std values are', df['balance'].mean(), ' and ', df['balance'].std())
```

```python
#6. Pick field *subscribed*, and label 'Yes' as 1, and 'No' as 0.

from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
encoder1.fit(df['subscribed'])
df['subscribed']=encoder1.transform(df['subscribed'])

display(df)
print(f'The new values are', df['subscribed'].unique())
```

```python
#7. Pick field *marital*, and do one hot encoding. Remove the original column.

print(df['marital'].unique())
df = pd.get_dummies(df, columns=['marital'],drop_first=True) #Remove first level to get k-1 dummies out of k categorical levels.

display(df)
```

```python
#8. Pick fields *default, housing, loan*, and do one hot encoding. 
# Remove the original column and remove the first encoded column.

df = pd.get_dummies(df, columns=['default','housing','loan'],drop_first=True)

display(df)
```

```python
#9. Pick field job, and discretize the categories as follows:
#     'unemployed':'C1', 'self-employed':'C1', 'student':'C1', 'retired':'C1'
#     'services':'C2', 'blue-collar':'C2', 'technician':'C2', 'housemaid':'C2'
#     'management':'C3', 'entrepreneur':'C3', 'admin.':'C3'
#     delete all records with 'unknown' value.
#     Then do one hot encoding and remove the original column.

# row indicies which has unknown value for education
index_row = df[ df['job'] == 'unknown' ].index

# Delete these row indexes from dataFrame
df.drop(index_row , inplace=True)
df['job'].unique().tolist()
```

```python
# data conversion as per requirements
job_mapping = {'unemployed':'C1', 'self-employed':'C1', 'student':'C1', 'retired':'C1',
              'services':'C2', 'blue-collar':'C2', 'technician':'C2', 'housemaid':'C2',
              'management':'C3', 'entrepreneur':'C3', 'admin.':'C3'}

df['job'] = df['job'].map(job_mapping)   # note: you can also use custom function and apply method.
display(df)
```

```python
# one hot encoding
df = pd.get_dummies(df, columns=['job'],drop_first=True)

#10. Display and save the transformed dataset 

# Display the DF
display(df)
```

```python

# Save the DF as "Dataset-2-Clean.csv" in data folder

df.to_csv('data/Dataset-2-Clean.csv', index = False, header=True)
```
